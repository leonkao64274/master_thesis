```python
import pandas as pd
import numpy as np
```

<font size=6> Performance

```python
def evaluate2(pred,real):  
    tp=0
    tn=0
    fp=0
    fn=0
    for i in range(len(pred)):
        if pred[i]==0:
            if real[i]==1:
                fp+=1
            else:
                tp+=1
        else:
            if real[i]==1:
                tn+=1
            else:
                fn+=1
    accuracy=round((tp+tn)/(tn+tp+fn+fp),3)
    precision=round(tp/(tp+fp),3)
    recall=round(tp/(tp+fn),3)
    mcc=round((tp*tn-fp*fn)/((tp+fp)*(tp+fn)*(tn+fp)*(tn+fn))**(1/2),3)
    f1=round((2*(precision*recall)/(precision+recall)),3)
    return accuracy,precision,recall,mcc,f1
```

<font size=6> nonTATA Cross Prediction & number of data

```python
allp=[]
allr=[]
allf=[]
for i in range(1,7):
    precisions=[]
    recalls=[]
    f1s=[]
    for j in range(1,7):
        accuracy,precision,recall,mcc,f1=evaluate2(pred=np.load("models/ver4/pred_"+str(i)+"p"+str(j)+"_nonTATA.npy"),\
                                                   real=np.load("nonTATAlabels/nT"+str(j)+"labels.npy"))
        precisions.append(precision);recalls.append(recall);f1s.append(f1)
    allp.append(precisions);allr.append(recalls);allf.append(f1s)
table_allp = pd.DataFrame(allp, columns=["mouse model", "shrew model", "rat model", "rabbit model","human model","dog model"],\
                          index=["mouse", "shrew", "rat", "rabbit","human","dog"])
table_allr = pd.DataFrame(allr, columns=["mouse model", "shrew model", "rat model", "rabbit model","human model","dog model"],\
                          index=["mouse", "shrew", "rat", "rabbit","human","dog"])
table_allf = pd.DataFrame(allf, columns=["mouse model", "shrew model", "rat model", "rabbit model","human model","dog model"],\
                          index=["mouse", "shrew", "rat", "rabbit","human","dog"])
#------------------------------------------------------------
real=np.load("nonTATAlabels/nT6labels.npy")
for i in range(len(real)-1):
    if real[i]!=real[i+1]:
        dog=[i,len(real)-i,len(real)]
real=np.load("nonTATAlabels/nT5labels.npy")
for i in range(len(real)-1):
    if real[i]!=real[i+1]:
        human=[i,len(real)-i,len(real)]
real=np.load("nonTATAlabels/nT4labels.npy")
for i in range(len(real)-1):
    if real[i]!=real[i+1]:
        rabbit=[i,len(real)-i,len(real)]
real=np.load("nonTATAlabels/nT3labels.npy")
for i in range(len(real)-1):
    if real[i]!=real[i+1]:
        rat=[i,len(real)-i,len(real)]
real=np.load("nonTATAlabels/nT2labels.npy")
for i in range(len(real)-1):
    if real[i]!=real[i+1]:
        shrew=[i,len(real)-i,len(real)]
real=np.load("nonTATAlabels/nT1labels.npy")
for i in range(len(real)-1):
    if real[i]!=real[i+1]:
        mouse=[i,len(real)-i,len(real)]
data_num = pd.DataFrame([mouse,shrew,rat,rabbit,human,dog], columns=["TSS", "nonTSS", "total"],\
                        index=["mouse", "shrew", "rat", "rabbit","human","dog"])
```

<font size=6> Homolog

```python
def homolog(): 
    rdata=[]
    with open ("homologene.data","r") as f:
        rdata=f.readlines()
    split_data=[]
    for i in range(len(rdata)):
        split_data.append(rdata[i].split())
#    split_data
    species={"10090":"mouse","10116":"rat","9615":"dog","9606":"human"}
    fourdata=[]
    for i in range(len(split_data)):
        if split_data[i][1] in species.keys():
            split_data[i][1]=species[split_data[i][1]]
            fourdata.append(split_data[i])
#    fourdata
    mr=0;mh=0;md=0;rh=0;rd=0;hd=0
    tmp=fourdata[0][0]
    p1=0
    p2=0
    tmplist=[]
    for i in range(len(fourdata)):
        if fourdata[i][0]!=tmp:
            p1=p2
            p2=i
            tmp=fourdata[i][0]
#            print(p1,p2)
            for j in range(p1,p2):
                tmplist.append(fourdata[j][1])
            if ("mouse" in tmplist) and ("rat" in tmplist):
                mr+=1
            if ("mouse" in tmplist) and ("human" in tmplist):
                mh+=1
            if ("mouse" in tmplist) and ("dog" in tmplist):
                md+=1
            if ("rat" in tmplist) and ("human" in tmplist):
                rh+=1
            if ("rat" in tmplist) and ("dog" in tmplist):
                rd+=1
            if ("human" in tmplist) and ("dog" in tmplist):
                hd+=1
#            print(tmplist)
            tmplist=[]
    return mr,mh,md,rh,rd,hd
mr,mh,md,rh,rd,hd=homolog()
homolog_data = pd.DataFrame([["\\",mr,mh,md],[mr,"\\",rh,rd],[mh,rh,"\\",hd],[md,rd,hd,"\\"]],\
                            columns=["mouse", "rat", "human","dog"],index=["mouse","rat","human","dog"])
```

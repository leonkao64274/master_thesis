<font size=6> function of load and choose

```python
import numpy as np
```

```python
T1=np.load("TATA/T1.npy")
T2=np.load("TATA/T22.npy") 
T3=np.load("TATA/T3.npy") 
T4=np.load("TATA/T44.npy") 
T5=np.load("TATA/T5.npy") 
T6=np.load("TATA/T6.npy") 
#------------------------------------
T1labels=np.load("TATAlabels/T1labels.npy")
T2labels=np.load("TATAlabels/T22labels.npy")
T3labels=np.load("TATAlabels/T3labels.npy")
T4labels=np.load("TATAlabels/T44labels.npy") 
T5labels=np.load("TATAlabels/T5labels.npy")
T6labels=np.load("TATAlabels/T6labels.npy")
#------------------------------------
nT1=np.load("nonTATA/nT1.npy")
nT2=np.load("nonTATA/nT22.npy") 
nT3=np.load("nonTATA/nT3.npy") 
nT4=np.load("nonTATA/nT44.npy") 
nT5=np.load("nonTATA/nT5.npy") 
nT6=np.load("nonTATA/nT6.npy") 
#------------------------------------
nT1labels=np.load("nonTATAlabels/nT1labels.npy")
nT2labels=np.load("nonTATAlabels/nT22labels.npy")
nT3labels=np.load("nonTATAlabels/nT3labels.npy")
nT4labels=np.load("nonTATAlabels/nT44labels.npy") 
nT5labels=np.load("nonTATAlabels/nT5labels.npy")
nT6labels=np.load("nonTATAlabels/nT6labels.npy")
datalabels=[0,[T1,nT1,T1labels,nT1labels],[T2,nT2,T2labels,nT2labels],[T3,nT3,T3labels,nT3labels],\
              [T4,nT4,T4labels,nT4labels],[T5,nT5,T5labels,nT5labels],[T6,nT6,T6labels,nT6labels]]
```

```python
def choose(*key,place,name):
    TATA=[]
    nonTATA=[]
    TATAlabels=[]
    nonTATAlabels=[]
    for i in range(len(key)):
        TATA+=list(datalabels[key[i]][0])
        nonTATA+=list(datalabels[key[i]][1])
        TATAlabels+=list(datalabels[key[i]][2])
        nonTATAlabels+=list(datalabels[key[i]][3])
    TATA=np.array(TATA)
    nonTATA=np.array(nonTATA)
    TATAlabels=np.array(TATAlabels)
    nonTATAlabels=np.array(nonTATAlabels)
    np.save(("TATA/"+place+"/"+name+"TATA"),TATA)
    np.save(("nonTATA/"+place+"/"+name+"nonTATA"),nonTATA)
    np.save(("TATAlabels/"+place+"/"+name+"TATAlabels"),TATAlabels)
    np.save(("nonTATAlabels/"+place+"/"+name+"nonTATAlabels"),nonTATAlabels)
    
    return 

#     return TATA,nonTATA,TATAlabels,nonTATAlabels
```

```python
from tqdm import tqdm
```

<font size=5>get gene info 

```python
def orthoid2geneid():    
    rdata=[]
    with open("odb10v1_genes.tab","r") as f:
        rdata=f.readlines()
    rgene_info=[line.split("\t")for line in tqdm(rdata)]
    gene_info=[[sublist[0],sublist[1],sublist[6]] for sublist in tqdm(rgene_info)]
    return gene_info
gene_info=orthoid2geneid()
```

```python
gene_info=[sublist for sublist in gene_info if (sublist[0][0:4]=="9606" or sublist[0][0:5]=="10090"or sublist[0][0:5]=="10116")]
```

<font size=5> gene to og

```python
def gene2og():
    rdata2=[]
    with open ("odb10v1_OG2genes.tab","r") as f:
        rdata2=f.readlines()
    ogs=[line.split("\t")[0] for line in tqdm(rdata2)]
    orthoids_2=[line.split("\t")[1] for line in tqdm(rdata2)]
    homo_ogs_orthoids=[[ogs[i],orthoids_2[i].rstrip()] for i in tqdm(range(len(orthoids_2))) if orthoids_2[i][0:4]=="9606"]
    mouse_ogs_orthoids=[[ogs[i],orthoids_2[i].rstrip()] for i in tqdm(range(len(orthoids_2))) if orthoids_2[i][0:5]=="10090"]
    rat_ogs_orthoids=[[ogs[i],orthoids_2[i].rstrip()] for i in tqdm(range(len(orthoids_2))) if orthoids_2[i][0:5]=="10116"]
    return homo_ogs_orthoids,mouse_ogs_orthoids,rat_ogs_orthoids
homo_ogs_orthoids,mouse_ogs_orthoids,rat_ogs_orthoids=gene2og()
```

<font size=5> rmh og and from orthoid to gene id

```python
def rmh_m_og():
    homo_ogs=[sublist[0] for sublist in homo_ogs_orthoids]
    mh_ogs_mouseorthoids=[sublist for sublist in tqdm(mouse_ogs_orthoids) if sublist[0] in homo_ogs]
    rat_ogs=[sublist[0] for sublist in rat_ogs_orthoids]
    rmh_ogs_mouseorthoids=[sublist for sublist in tqdm(mh_ogs_mouseorthoids) if sublist[0] in rat_ogs]
    rmh_mousegenes=[sublist[1] for sublist in rmh_ogs_mouseorthoids]
    mouse_geneids=[sublist[2] for sublist in tqdm(gene_info)  if sublist[0][0:5]=="10090" and sublist[0] in rmh_mousegenes ]
    return mouse_geneids
mouse_geneids=rmh_m_og()
with open("rmh_m_geneid.txt","w") as f:
    for sublist in mouse_geneids:
        f.write(sublist+"\n")
```

```python
def rmh_h_og():
    mouse_ogs=[sublist[0] for sublist in mouse_ogs_orthoids]
    mh_ogs_homoorthoids=[sublist for sublist in tqdm(homo_ogs_orthoids) if sublist[0] in mouse_ogs]
    rat_ogs=[sublist[0] for sublist in rat_ogs_orthoids]
    rmh_ogs_homoorthoids=[sublist for sublist in tqdm(mh_ogs_homoorthoids) if sublist[0] in rat_ogs]
    rmh_homogenes=[sublist[1] for sublist in rmh_ogs_homoorthoids]
    homo_geneids=[sublist[2] for sublist in tqdm(gene_info)  if sublist[0][0:4]=="9606" and sublist[0] in rmh_homogenes ]
    return homo_geneids
homo_geneids=rmh_h_og()
with open("rmh_h_geneid.txt","w") as f:
    for sublist in homo_geneids:
        f.write(sublist+"\n")
```

```python
def rmh_r_og():
    mouse_ogs=[sublist[0] for sublist in mouse_ogs_orthoids]
    mh_ogs_ratorthoids=[sublist for sublist in tqdm(rat_ogs_orthoids) if sublist[0] in mouse_ogs]
    homo_ogs=[sublist[0] for sublist in homo_ogs_orthoids]
    rmh_ogs_ratorthoids=[sublist for sublist in tqdm(mh_ogs_ratorthoids) if sublist[0] in homo_ogs]
    rmh_ratgenes=[sublist[1] for sublist in rmh_ogs_ratorthoids]
    rat_geneids=[sublist[2] for sublist in tqdm(gene_info)  if sublist[0][0:5]=="10116" and sublist[0] in rmh_ratgenes ]
    return rat_geneids
rat_geneids=rmh_r_og()
with open("rmh_r_geneid.txt","w") as f:
    for sublist in rat_geneids:
        f.write(sublist+"\n")
```

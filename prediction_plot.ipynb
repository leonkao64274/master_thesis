```python

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def plot_npy(fname):
    y=np.load(fname)[15000:25000]
    x=range(-int((len(y))/2)+1200,int((len(y))/2)+1200)
    plt.bar(x,y)
    plt.show()
    

#histogream 和 bar chart 不同
```
